```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import aiohttp
import time
import requests, os, time, random
from News.ip_rotator import ApiGateway
import creds
from binanceVision import BinanceDownloader
from bs4 import BeautifulSoup
```

```python
'''  
We are interested in the ones that are newly traded.
- 우선 Spot만 있어도 된다. Doesn't have to be futures



'''
```

```python
gateway = ApiGateway("https://www.binance.com",
                     regions=["ap-northeast-1"],
                     access_key_id=creds.aws_key, access_key_secret=creds.aws_secret,)  # Use the region you set up the API in
gateway.start()
session = requests.Session()
session.mount("https://", gateway)
```

```python
params = {"type": "1","pageSize": "50", "pageNo": "1", "catalogId":"48"}

news_list = []
for pageNo in np.arange(1,30):
    params['pageNo'] = pageNo
    response = session.get("https://www.binance.com/bapi/composite/v1/public/cms/article/list/query", params=params, headers ={})
    try:
        articles = response.json()['data']['catalogs'][0]['articles']
        for article in articles:
            news = {'exchange':'binance', 'title':article['title'], 'date': pd.to_datetime(article['releaseDate'],unit='ms', utc=True),
                    'link': f"https://www.binance.com/en/support/announcement/{article['code']}"}
            news_list.append(news)
    except:
        continue
```
